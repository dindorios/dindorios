```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Carregar a imagem
imagem = cv2.imread('/content/Imagemparaoprojeto.png)

```

```python
# Verificar se a imagem foi carregada corretamente
if imagem is None:
    print("Erro ao carregar a imagem. Verifique o caminho.")
else:
    # Converter para escala de cinza
    cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

    # Binarização (usando limiarização simples)
    limiar = 127  # Ajuste este valor conforme necessário
    _, binaria = cv2.threshold(cinza, limiar, 255, cv2.THRESH_BINARY)

    # Exibir imagens usando matplotlib
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))  # Converter BGR para RGB
    plt.title('Imagem Original')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(cinza, cmap='gray')
    plt.title('Imagem em Escala de Cinza')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(binaria, cmap='gray')
    plt.title('Imagem Binária')
    plt.axis('off')

    plt.show()
```
