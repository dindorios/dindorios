```python
# Importação de Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Definindo a matriz de confusão
VP = 50  # Verdadeiros Positivos
FN = 10  # Falsos Negativos
FP = 5   # Falsos Positivos
VN = 35  # Verdadeiros Negativos
```

```python
# Função para calcular as métricas
def calcular_metricas(VP, FN, FP, VN):
    # Verificação para evitar divisão por zero
    if (VP + VN + FP + FN) == 0:
        return "Erro: Total de elementos não pode ser zero."

    # Acurácia
    acuracia = (VP + VN) / (VP + VN + FP + FN)

    # Sensibilidade (Recall)
    sensibilidade = VP / (VP + FN) if (VP + FN) > 0 else 0

    # Especificidade
    especificidade = VN / (VN + FP) if (VN + FP) > 0 else 0

    # Precisão
    precisao = VP / (VP + FP) if (VP + FP) > 0 else 0

    # F-score
    F_score = 2 * (precisao * sensibilidade) / (precisao + sensibilidade) if (precisao + sensibilidade) > 0 else 0

    return acuracia, sensibilidade, especificidade, precisao, F_score
```

```python
# Executando o cálculo
metricas = calcular_metricas(VP, FN, FP, VN)
```

```python
# Criando a matriz de confusão
matriz_confusao = np.array([[VP, FN],
                             [FP, VN]])
```

```python
# Exibindo a matriz de confusão usando Matplotlib
plt.figure(figsize=(8, 6))
sns.heatmap(matriz_confusao, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predito Positivo', 'Predito Negativo'],
            yticklabels=['Real Positivo', 'Real Negativo'])
plt.title('Matriz de Confusão')
plt.ylabel('Classe Real')
plt.xlabel('Classe Predita')
plt.show()
```

```python
# Exibindo os resultados
if isinstance(metricas, tuple):
    print(f"Acurácia: {metricas[0]:.2f}")
    print(f"Sensibilidade (Recall): {metricas[1]:.2f}")
    print(f"Especificidade: {metricas[2]:.2f}")
    print(f"Precisão: {metricas[3]:.2f}")
    print(f"F-score: {metricas[4]:.2f}")
else:
    print(metricas)  # Exibe a mensagem de erro
    print(metricas)  # Exibe a mensagem de erro
```
